```python
import requests

def get_quote():
    url = "https://api.forismatic.com/api/1.0/" #API Forismatic предоставляет программный доступ к коллекции вдохновляющих цитат
                                                #Посредством простых HTTP-запросов GET или POST
                                                #Это бесплатный общедоступный API, простой в использовании для разработчиков
    params = {
        "lang": "ru",        # выбираем русский
        "method": "getQuote",
        "format": "json"
    }
    try:
        response = requests.get(url, params=params, timeout=5)
        response.raise_for_status()  # проверка
        data = response.json()
        print(data)
        quote = data.get("quoteText")
        author = data.get("quoteAuthor") or "Неизвестный"
        return f"Цитата: {quote}\nАвтор: {author}"
    except Exception as e:
        return f"Ошибка: {e}"
def main():
    print("Генератор цитат:")
    print("Команды:")
    print("1 — новая цитата")
    print("2 — завершение программы")
    print("'помощь' — показать это меню")
    while True:
        user_input = input("Введите команду  ").strip().lower()

        if user_input in ['2','22','222']:
            print("До свидания\n")
            break
        if user_input == 'помощь':
            print("\n меню :")
            print("  • 1 → новая цитата")
            print("  • 2 → выход")
            print("  • помощь → это меню\n")
            continue
        if user_input in ['1','11','111'] :
            # Получение
            result = get_quote()
            print(result)



if __name__ == "__main__":
    main()
```
